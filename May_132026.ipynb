```python
users = []

while True:
    username = input("Enter Username (or type stop): ")

    if username == "stop":
        break

    password = input("Enter Password: ")

    users.append({
        "username": username,
        "password": password
    })

print("\nSaved Users:")
for user in users:
    print(user)
```
