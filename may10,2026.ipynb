```python
word = "kavitha"
count = 0
for each in word:
  if each == "a":
    count=count+1
  elif  each == "e":
    count=count+1
  elif each == "i":
    count=count+1
  elif each == "o":
    count=count+1
  elif each == "u":
    count=count+1

print(count)

```
